```python
# Total unique customers

# Table Representation
#removing duplicates from customer id column
unique_customers=df.drop_duplicates(subset=['Customer ID'])

# male customers count
male_customers=unique_customers[unique_customers['Gender']=='Male']['Customer ID'].count()

# female customers count
female_customers=unique_customers[unique_customers['Gender']=='Female']['Customer ID'].count()

#total unique customers
total_customers=male_customers+female_customers

#customers distribution according to their gender
customer_data=pd.DataFrame(pd.DataFrame({
    'Customers': ['Total Unique Customers','Male', 'Female'],
    'Count': [male_customers+female_customers,male_customers, female_customers]
}))

# Visual Representation
barplot=sns.countplot(x='Gender',data=unique_customers,hue='Gender',palette='Accent',width=0.65,gap=0.01)
plt.title("Customer Gender Distribution Data")
plt.ylabel("Count of Customers")
for i in barplot.containers:
    barplot.bar_label(i)
plt.legend(loc='upper right', title='Gender')
plt.show()
```

```python
# Removed Duplicates from Customer ID
unique_df = df.drop_duplicates(subset=['Customer ID'])

#age-group arraya
age_group=[(18,30),(31,50),(51,70),(71,90)]

#total male customers array according to age-group
male_customer=[]

#total female customers array according to age-group
female_customer=[]

for age1,age2 in age_group:

    data=unique_df[
        (unique_df['Age'] >= age1) & 
        (unique_df['Age'] <= age2)
        ]['Customer ID'].count()

    data1=unique_df[
        (unique_df['Gender'] == 'Male') & 
        (unique_df['Age'] >= age1) & 
        (unique_df['Age'] <= age2)
        ]['Customer ID'].count()

    data2=unique_df[
        (unique_df['Gender'] == 'Female') & 
        (unique_df['Age'] >= age1) & 
        (unique_df['Age'] <= age2)
        ]['Customer ID'].count()

    male_customer.append(data1)
    female_customer.append(data2)

#customers distribution data according to different age-groups
age_gender_data=pd.DataFrame({
    'Age Group': ['18-30', '31-50', '51-70', 'Above 70'],
    'Male Customers': male_customer,
    'Female Customers': female_customer
});


new_age_gender_data = age_gender_data.melt(id_vars='Age Group')
bar_plot=sns.barplot(x='Age Group',y='value',data=new_age_gender_data,hue='variable',palette='Accent')
plt.title("Age Gender Customer Distribution")
plt.ylabel("Count of Customers")
for i in bar_plot.containers:
    bar_plot.bar_label(i)
plt.show()
```

```python
# Number of completed orders
completed_order = df[df['Order Status'] == 'Completed']['Customer ID'].count()

# Number of cancelled orders
cancelled_order = df[df['Order Status'] == 'Cancelled']['Customer ID'].count()

# Order Status Distribution
order_status_dist = pd.DataFrame({
    'Orders': ['Completed Orders','Cancelled Orders'],
    'Count':[completed_order,cancelled_order]
})
a=[0.03,0]
plot=plt.pie(order_status_dist['Count'],labels=order_status_dist['Orders'],data='order_status_dist',autopct="%0.2f%%",radius=0.8,textprops={"fontsize": 9},explode=a,startangle=30,wedgeprops={'edgecolor':"white"},colors=['lightgreen','lightblue']
)
plt.legend(loc=2)
plt.title("Total Orders Distribution")
plt.show()
```

```python
# array of cancelled orders according to different age-groups
cancelled_orders=[]

# array of completed orders according to different age-groups
completed_orders=[]

for age1,age2 in age_group:

    data1=df[(df['Order Status']=='Completed') & 
            (unique_df['Age'] >= age1) & 
            (unique_df['Age'] <= age2)         
            ]['Customer ID'].count()

    data2=df[(df['Order Status']=='Cancelled') & 
            (unique_df['Age'] >= age1) & 
            (unique_df['Age'] <= age2)         
            ]['Customer ID'].count()

    completed_orders.append(data1)
    cancelled_orders.append(data2)

#Order Status data according to different age-groups
result=pd.DataFrame({
    'Age Group': ['18-30', '31-50', '51-70', 'Above 70'],
    'Completed Orders':completed_orders,
    'Cancelled Orders': cancelled_orders
})
new_result=result.melt(id_vars='Age Group')

new_result_plot=sns.barplot(x='Age Group',y='value',data=new_result,hue='variable',palette='Set2')
plt.title("Completed and Cancelled Orders according to different Age Groups")
for i in new_result_plot.containers:
    new_result_plot.bar_label(i)
plt.ylabel("Count of Orders")
plt.show()
```

```python
# calculating table where order status is completed
order_status_completed=df[df['Order Status']=='Completed']

# calculating table where order status is cancelled
order_status_cancelled=df[df['Order Status']=='Cancelled']

#calculating completed orders 
completed=order_status_completed.groupby('Product Type')['Customer ID'].count()

#calculating completed orders
cancelled=order_status_cancelled.groupby('Product Type')['Customer ID'].count()

#cancelled and completed orders data of each product type
product_order_summary = pd.DataFrame({
    'Completed Orders': completed,
    'Cancelled Orders': cancelled
})

product_order_summary = product_order_summary.reset_index()

new_product_order_summary = product_order_summary.melt(id_vars='Product Type')
product_order_plt=sns.barplot(x='Product Type',y='value',data=new_product_order_summary,hue='variable',palette='Set2')
plt.title("Completed and Cancelled Orders Of Each Product Type")
plt.ylabel("Count of Order")
for i in product_order_plt.containers:
    product_order_plt.bar_label(i)
plt.show()
```

```python
#Table Representation 
count_payment_method=order_status_completed.groupby('Payment Method')['Customer ID'].count();

payment_method_dist=pd.DataFrame({
    'Count':count_payment_method
})
payment_method_dist.rename(columns={"Count":"Number of Payments Completed"})

# Data Visualization
payment_method_data=sns.countplot(x='Payment Method',data=order_status_completed,hue='Payment Method',palette='Set3',gap=0.2,width=1)
plt.title("Completed Payments by Different Payment Methods")
plt.ylabel("Number of Payments Completed")
for i in payment_method_data.containers:
    payment_method_data.bar_label(i)
plt.show()
```

```python
product_type_payment_data=order_status_completed.groupby(['Product Type','Payment Method'])['Customer ID'].count()

product_type_payment_method = product_type_payment_data.unstack(fill_value=0)

product_payment=sns.countplot(x='Product Type',data=order_status_completed,hue='Payment Method',palette='Set3',width=0.7)
plt.title("Count of Payments Methods for each Product Type")
plt.ylabel("Count of Payments Done")

plt.show()
```

```python
# Mean Rating of each product type where order status is completed 
avg_rating_each_product = order_status_completed.groupby("Product Type")[['Rating']].mean().sort_values(by='Rating',ascending=False)

# Rounding the rating of each product to 2 decimal places
avg_rating_each_product['Rating']=avg_rating_each_product['Rating'].round(2)

avg_rating_each_product = avg_rating_each_product.rename(columns={"Rating": "Average Rating"}).reset_index()

product_rating = sns.barplot(x='Product Type',y='Average Rating',data=avg_rating_each_product,hue='Product Type',palette='Set2')
plt.title("Average Rating of Each Product Type")
for i in product_rating.containers:
    product_rating.bar_label(i)
plt.show()
```

```python
# total product price of each product type
product_price=order_status_completed.groupby('Product Type')['Total Price'].sum()

# total add-on price of each product type
addon_price=order_status_completed.groupby('Product Type')['Add-on Total'].sum()

# unit price of each product type
unit_price=order_status_completed.groupby('Product Type')['Unit Price'].mean()

#Complete Revenue Distribution of each Product Type
product_rev=pd.DataFrame({
    'Prodcut Revenue':product_price,
    'Add-on Purchase Revenue':addon_price,
    'Unit Price':unit_price,
    'Total Revenue':product_price+addon_price
})
product_rev=product_rev.reset_index()
product_rev

new_product_rev=product_rev.melt(id_vars='Product Type')
product_rev=sns.barplot(x='Product Type',y='value',data=new_product_rev,hue='variable',palette='Set2')
plt.title("Revenue Obtained from Each Product Type")
plt.ylabel("Revenue")
plt.ylim(0,1e7)
plt.show()
```

```python
#group by product type and shipping type
product_shipping=order_status_completed.groupby(['Product Type','Shipping Type'])['Customer ID'].count()

#making table of complete data of count of each shipping type for each product type
product_shipping_data=product_shipping.unstack(fill_value=0)
product_shipping_data

product_shipping = sns.countplot(x='Product Type',data=order_status_completed,hue='Shipping Type',palette='Set3',width=0.7)
plt.title("Count of Shipping Type for each Product Type")
plt.ylabel("Count of Shipping Types")

plt.show()
```

```python
# Merge product order summary with payment method data
merged_data = pd.merge(product_order_summary, product_type_payment_method, on='Product Type')

# Merge with average rating for each product
merged_data = pd.merge(merged_data, avg_rating_each_product, on='Product Type')

#  Rename 'Rating' column to 'Average Rating'
merged_data.rename(columns={'Rating': 'Average Rating'}, inplace=True)

#  Merge with product revenue data
merged_data = pd.merge(merged_data, product_rev, on='Product Type')

#  Merge with shipping data
final_data = pd.merge(merged_data, product_shipping_data, on='Product Type')
final_data
```

```python
# For loyalty customers
rev_loyalty_customer = new_df[(new_df['Order Status'] == 'Completed') & (new_df['Loyalty Member'] == 'Yes')]
loyalty_customer_revenues = rev_loyalty_customer[['Total Price', 'Add-on Total']].sum()

# For non-loyalty customers
rev_nonloyalty_customer = new_df[(new_df['Order Status'] == 'Completed') & (new_df['Loyalty Member'] == 'No')]
nonloyalty_customer_revenues = rev_nonloyalty_customer[['Total Price', 'Add-on Total']].sum()

# Calculating total revenue for both groups
loyalty_customer_total_rev = loyalty_customer_revenues['Total Price'] + loyalty_customer_revenues['Add-on Total']
nonloyalty_customer_total_rev = nonloyalty_customer_revenues['Total Price'] + nonloyalty_customer_revenues['Add-on Total']

# Creating DataFrame
rev_customer_dist = pd.DataFrame({
    'Customers': ['Loyalty', 'Non-Loyalty'],
    'Products Revenue': [loyalty_customer_revenues['Total Price'], nonloyalty_customer_revenues['Total Price']],
    'Add-on Product Revenue': [loyalty_customer_revenues['Add-on Total'], nonloyalty_customer_revenues['Add-on Total']],
    'Total Revenue': [loyalty_customer_total_rev, nonloyalty_customer_total_rev]
})
a=[0.07,0]
plt.pie(rev_customer_dist['Total Revenue'],labels=rev_customer_dist['Customers'],autopct="%0.2f%%",radius=0.8,textprops={"fontsize": 9},explode=a,startangle=0,wedgeprops={'edgecolor':"white"},colors=['lightblue','lightgreen'])
plt.title("Revenue from Loyalty Members and Non Loyalty Members")
plt.ylabel("Revenue")
plt.legend()
plt.show()
```

```python
# for number of male loyal customers of different age-groups
male_loyal_customers=[]

# for number of female loyal customers of different age-groups
female_loyal_customers=[]

# loop for calculating data
for age1,age2 in age_group:

    male=unique_new_df[
    (unique_new_df['Age'] >= age1) & 
    (unique_new_df['Age'] <= age2) &
    (unique_new_df['Loyalty Member']=='Yes') & 
    (unique_new_df['Gender']=='Male')
    ]['Customer ID'].count()

    female=unique_new_df[
    (unique_new_df['Age'] >= age1) & 
    (unique_new_df['Age'] <= age2) &
    (unique_new_df['Loyalty Member']=='Yes') & 
    (unique_new_df['Gender']=='Female')
    ]['Customer ID'].count()

    male_loyal_customers.append(male)
    female_loyal_customers.append(female)

# Number of loyal customers according to gender and age-groups
loyalty_customer_dist=pd.DataFrame({
    'Age Group':['18-30','31-50','51-70','Above 70'],
    'Male Loyalty Customers':male_loyal_customers,
    'Female Loyalty Customers' : female_loyal_customers
})

new_loyalty_customer_dist = loyalty_customer_dist.melt(id_vars='Age Group')
loyalty_members=sns.barplot(x='Age Group',y='value',data=new_loyalty_customer_dist,hue='variable',palette='Set2')
plt.title("Age Gender Distribution of Loyalty Members")
plt.ylabel("Count of Customers")
for i in loyalty_members.containers:
    loyalty_members.bar_label(i)
plt.show()
```

```python
# Revenue from male loyal customers of different age-groups
male_rev=[]

# Revenue from female loyal customers of different age-groups
female_rev=[]

# loop for calculating revenue
for age1,age2 in age_group:
    male=new_df[(new_df['Order Status']=='Completed') & (new_df['Loyalty Member']=='Yes') & (new_df['Age'] >= age1) & (new_df['Age'] <= age2) & (new_df['Gender']=='Male')][['Total Price', 'Add-on Total']].sum()
    female=new_df[(new_df['Order Status']=='Completed') & (new_df['Loyalty Member']=='Yes') & (new_df['Age'] >= age1) & (new_df['Age'] <= age2) & (new_df['Gender']=='Female')][['Total Price','Add-on Total']].sum()

    total_male_revenue=male['Total Price'] + male['Add-on Total']
    total_female_revenue=female['Total Price'] + female['Add-on Total']

    male_rev.append(total_male_revenue)
    female_rev.append(total_female_revenue)

# revenue distribution of loyal customers according to age and gender 
result=pd.DataFrame({
    'Age Group':['18-30','31-50','51-70','Above 70'],
    'Male Loyal Customers Revenue':male_rev,
    'Female Loyal Customers Revenue':female_rev
})
result1=result.melt(id_vars='Age Group')
rev_customers=sns.barplot(x='Age Group',y='value',data=result1,hue='variable',palette='Set2')
plt.title("Revenue from Loyalty Members")
plt.ylabel("Revenue")
for i in rev_customers.containers:
    rev_customers.bar_label(i)
plt.show()
```
